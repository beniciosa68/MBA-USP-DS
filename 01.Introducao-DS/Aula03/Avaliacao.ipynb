# Exercicio 2

Usando a distância interquantil e o critério de que uma observação é um outlier se ao menos uma das variáveis está fora dos limites máximos do boxplot, quantos outliers há nos dados da Iris? (Veja o exemplo da aula)

Escolha uma:<br>
<b>a. 4</b><br>
<b>b. 2</b><br>
<b>c. 7</b><br>
<b>d. 1</b><br>
<b>e. 10</b><br>

```python
import pandas as pd 

data = pd.read_csv('data/iris.csv', header=(0))

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

outlier = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)

num_outliers = [n for n in outlier if n is True]

print ("Resposta:",len(num_outliers))
```

# Exercicio 3

Usando a distância interquantil e o critério de que uma observação é um outlier se ao menos uma das variáveis está fora dos limites máximos do boxplot, quantos outliers há nos dados BostonHousing?

```python
import pandas as pd 

data = pd.read_csv('data/BostonHousing.csv', header=(0))

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

outlier = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)

num_outliers = [n for n in outlier if n is True]

print ("Resposta:",len(num_outliers)," -> mais de 200")
```

# Exercicio 4

Quantas variáveis são necessárias para se explicar ao menos 90% da variância nos dados Vehicle?

Escolha uma:<br>
<b>a. 5</b><br>
<b>b. 10</b><br>
<b>c. 8</b><br>
<b>d. 1</b><br>
<b>e. 2</b><br>

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

data = pd.read_csv('data/Vehicle.csv', header=(0))

classes = np.unique(data[data.columns[-1]])

data = data.to_numpy()
nrow,ncol = data.shape
y = data[:,-1]
X = data[:,0:ncol-1]

scaler = StandardScaler().fit(X)
X = scaler.transform(X)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
aux = 0
plt.figure(figsize=(7,4))
for c in classes:
    nodes = np.where(y == c)
    # mostra os dados,considerando uma cor para cada classe
    plt.scatter(pca_result[nodes,0], pca_result[nodes,1], s=50, color = colors[aux], label = c)
    aux = aux + 1
plt.legend()
plt.xlabel("First component", fontsize=20)
plt.ylabel("Second component", fontsize=20)
plt.xticks(color='k', size=20)
plt.yticks(color='k', size=20)
plt.show(True)

pca = PCA().fit(X)
plt.figure(figsize=(7,4))
ncomp = np.arange(1, np.shape(X)[1]+1)

plt.plot(ncomp, np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.xlabel('number of components', fontsize=20)
plt.ylabel('cumulative explained variance', fontsize=20);
plt.xticks(color='k', size=20)
plt.yticks(color='k', size=20)
plt.grid(True)
plt.show(True)

dic = dict(zip(ncomp, list(map('{:.2f}'.format,np.cumsum(pca.explained_variance_ratio_)))))

resposta = [key for key,val in dic.items() if val == '0.91']

print("Resposta:", resposta[0])
```

# Exercicio 5

Quantas variáveis são necessárias para se explicar ao menos 80% da variância nos dados BostonHousing?

Escolha uma:<br>
<b>a. 5</b><br>
<b>b. 1</b><br>
<b>c. 10</b><br>
<b>d. 8</b><br>
<b>e. 2</b><br>

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

boston = pd.read_csv('data/BostonHousing.csv')

# Como temos 14 variaveis, podemos reduzir a dimensionalidade para qualquer valor abaixo disso
# Portanto, criamos uma lista com todas possibilidades para analizarmos qual é o minimo
# dimensoes que podemos reduzir ainda sendo possivel explicar sua variancia
ncomp = np.arange(1, len(boston.columns) + 1)

# Vamos padronizar os dados, de modo a evitar o efeito da escala dos atributos.
scaler = StandardScaler().fit(boston)
boston_transformed = scaler.transform(boston)

# instanciamos o PCA sem especificar o numero de componentes que desejamos
# em seguida ajustamos ao nosso conjunto de dados
pca = PCA().fit(boston_transformed)

# mostra a variância acumulada para todos os possiveis numeros de componentes
# Notamos que com apenas duas variaveis conseguimos explicar 95% da variancia dos dados
plt.figure(figsize=(8, 5))
plt.plot(ncomp, np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.xlabel('number of components', fontsize=20)
plt.ylabel('cumulative explained variance', fontsize=20);
plt.xticks(color='k', size=16)
plt.yticks(color='k', size=16)
plt.grid(True)
plt.show(True)

dic = dict(zip(ncomp, list(map('{:.2f}'.format,np.cumsum(pca.explained_variance_ratio_)))))

resposta = [key for key,val in dic.items() if val == '0.81']

print("Resposta:", resposta[0])
```
