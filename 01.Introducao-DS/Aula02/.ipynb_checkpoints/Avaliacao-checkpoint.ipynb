```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(101)

def function(x):
    y = np.cos(2 * np.pi * x)
    return y

#training
N_train = 10
x_train=np.linspace(0, 1,N_train)
y_train = function(x_train) + np.random.normal(0,0.1, N_train)

fig = plt.figure(figsize=(8, 6))
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=100, label="training data")

#test
N_test = 10
x_test=np.linspace(0, 1,N_test)
y_test = function(x_test) + np.random.normal(0,0.1, N_test)

#theoretical curves
xt=np.linspace(0,1,100)
yt = function(xt)
plt.plot(xt,yt, '-r', label="Theoretical curve")
plt.legend()
plt.show(True)

#Testando diferentes valores de M
##--------------------------------------------------------------------------------------------------
#base dos dados para fazer o plot de X
basePlotX = np.linspace(0, 1, 200)

#M=1
M1 = x_train.reshape(-1, 1)

#M=2 
poly2 = PolynomialFeatures(degree=2)
M2 = poly2.fit_transform(x_train.reshape(-1, 1))

#M=4 
poly4 = PolynomialFeatures(degree=4)
M4 = poly4.fit_transform(x_train.reshape(-1, 1))

#M=8
poly8 = PolynomialFeatures(degree=8)
M8 = poly8.fit_transform(x_train.reshape(-1, 1))

#M=9
poly9 = PolynomialFeatures(degree=9)
M9 = poly9.fit_transform(x_train.reshape(-1, 1))

##--------------------------------------------------------------------------------------------------


# usa regressão linear para ajustar o modelo
##--------------------------------------------------------------------------------------------------
#M1
modelM1 = linear_model.LinearRegression()
modelM1.fit(M1, y_train)

#M2
modelM2 = linear_model.LinearRegression()
modelM2.fit(M2, y_train)

#M4
modelM4 = linear_model.LinearRegression()
modelM4.fit(M4, y_train)

#M8
modelM8 = linear_model.LinearRegression()
modelM8.fit(M8, y_train)

#M9
modelM9 = linear_model.LinearRegression()
modelM9.fit(M9, y_train)
##--------------------------------------------------------------------------------------------------

# gera o conjunto de plot e preve os valores de y. 
#OBS: a funcao reshape vai "modelar" nosso vetor para que ele possa ser usado na funcao predict
##--------------------------------------------------------------------------------------------------
#M1
XplotM1 = basePlotX.reshape(-1, 1) 
Y_predM1 = modelM1.predict(XplotM1)
labelM1 = 'Linear; M=1'

#M2
XplotM2 = poly2.fit_transform(basePlotX.reshape(-1, 1))
Y_predM2 = modelM2.predict(XplotM2)
labelM2 = 'Poly 2; M=2'

#M4
XplotM4 = poly4.fit_transform(basePlotX.reshape(-1, 1))
Y_predM4 = modelM4.predict(XplotM4)
labelM4 = 'Poly 4; M=4'

#M8
XplotM8 = poly8.fit_transform(basePlotX.reshape(-1, 1))
Y_predM8 = modelM8.predict(XplotM8)
labelM8 = 'Poly 8; M=8'

#M9
XplotM9 = poly9.fit_transform(basePlotX.reshape(-1, 1))
Y_predM9 = modelM9.predict(XplotM9)
labelM9 = 'Poly 9; M=9'
##--------------------------------------------------------------------------------------------------


# mostra os resultados
##--------------------------------------------------------------------------------------------------
fig = plt.figure(figsize=(8, 6))

#M=1
plt.plot(basePlotX, Y_predM1, label=labelM1, color = 'red')

#M=2
plt.plot(basePlotX, Y_predM2, label=labelM2, color = 'purple')

#M=4
plt.plot(basePlotX, Y_predM4, label=labelM4, color = 'green')

#M=8
plt.plot(basePlotX, Y_predM8, label=labelM8, color = 'orange')

#M=9
plt.plot(basePlotX, Y_predM9, label=labelM9, color = 'yellow')

##--------------------------------------------------------------------------------------------------

plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=100, label="training data")
plt.legend()
plt.show(True)

print('Resposta: ' '4')
```

# Exercicio 4

Considerando a base Vehicle, encontre o melhor número de vizinhos. Qual é a acurácia aproximada usando validação cruzada com 10 folds, usando o melhor número de vizinhos encontrado? Repita o que foi feito na aula para a base iris.

Escolha uma:<br>

<b>a. 0.6<b/><br>
<b>b. 0.9<b/><br>
<b>c. 0.2<b/><br>
<b>d. 1.0<b/><br>
<b>e. 0.3<b/><br>

```python
import random
random.seed(1)
# Pandas é usado para manipulacao de dados
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do arquivo CSV. Lembre de baixar os dados disponibilizados no site do MBA.
# Na mesma pasta que você está usando esse jupyter notebook, deve ser extraido os dados baixados.
# Note que ele entra na pasta "data" e abre o arquivo "BreastCancer.csv". Então você deve ter essa configuração na pasta que está o seu jupyter notebook
data = pd.read_csv('data/Vehicle.csv', header=(0))
data = data.dropna(axis='rows') # Remove valores nulos
#print(data.shape)
data.head(25)

data = data.to_numpy()
nrow,ncol = data.shape
y = data[:,-1]
X = data[:,0:ncol-1]


from sklearn.model_selection import train_test_split

p = 0.2 # fracao de elementos no conjunto de teste
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = p)

from sklearn.model_selection import cross_validate
from sklearn.neighbors import KNeighborsClassifier

nkf = 10
vk = []
vscore = []
for k in range(1, 30):
    model = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')
    cv = cross_validate(model, x_train, y_train, cv=nkf)
    print('k:', k, 'accurace:', cv['test_score'].mean())
    vscore.append(cv['test_score'].mean())
    vk.append(k)

best_k = np.argmax(vscore)+1
print('Melhor k:', best_k)
plt.figure(figsize=(10,5))
plt.plot(vk, vscore, '-bo')
plt.xlabel('k', fontsize = 15)
plt.ylabel('Accuracy', fontsize = 15)
plt.show()

from sklearn.metrics import accuracy_score
model = KNeighborsClassifier(n_neighbors=best_k, metric = 'euclidean')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print('Accuracy:', accuracy_score(y_pred, y_test))
print()
print('Resposta:', '0.6')
```
