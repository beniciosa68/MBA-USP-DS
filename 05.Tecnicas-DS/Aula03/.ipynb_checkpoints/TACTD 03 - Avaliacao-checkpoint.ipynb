### Questão 3)

Normalize 2 variáveis da base: TOT e ENSINO MEDIO
* ENSINO MEDIO utilizando normalização da média
* TOT utilizando normalização L-2

Para isso, codifique funções que recebam uma coluna por parâmetro e retornem um atributo já normalizado

Depois, aplique as funções e crie novas variáveis com os atributos normalizados: MEDIO_nm e TOT_l2.

Após normalização, quais os valores de  média, desvio padrão, mínimo e máximo dessas variáveis, arredondando para 3 casas decimais?

<b>(a) MEDIO_nm: 0.000, 0.132, 0.000, 0.876; TOT_l2: 0.012, 0.007, 0.000, 1.000<br>
(b) MEDIO_nm: 0.000, 0.132, -0.124, 0.876; TOT_l2: 0.012, 0.007, 0.000, 0.043<br>
(c) MEDIO_nm: 1.000, 0.132, 0.000, 2.000; TOT_l2: 0.012, 0.007, 0.000, 1.000<br>
(d) MEDIO_nm: 0.000, 0.500, -1.124, 1.125; TOT_l2: 0.012, 0.007, 0.000, 0.043<br>



```python
print('Resposta: (b) MEDIO_nm: 0.000, 0.132, -0.124, 0.876; TOT_l2: 0.012, 0.007, 0.000, 0.043')
```

```python
#Normalização

#atributos
attrs = ['ENSINO MEDIO','TOT']
data.boxplot(attrs)
```

```python
#função normalização média
def nMedia(x,x_mean):
    return (x-x_mean)/(max(x)-min(x))

#função normalização L2
def nL2(x):
    return x/(np.sqrt(np.sum(x ** 2)))
```

```python
#aplicando a normalização para cada atributo
for var in attrs:
    
    #se for ENSINO MEDIO aplica normalização pela Média
    if(var == 'ENSINO MEDIO'):
        var_mean = data[var].mean()
        var_norm = nMedia(data[var],var_mean)
        newvar = 'MEDIO_NM'
        data[newvar] = var_norm
        
    #se for TOT aplica normalização L2    
    if(var == 'TOT'):    
        var_norm = nL2(data[var])
        newvar = 'TOT_L2'
        data[newvar] = var_norm

data.head()
```

```python
#imprimindo resultados
attrs_n = ['MEDIO_NM','TOT_L2']
print('MEDIO_NM:','%.3f' % data[attrs_n[0]].describe()[1],'%.3f' % data[attrs_n[0]].describe()[2],'%.3f' % data[attrs_n[0]].describe()[3],'%.3f' % data[attrs_n[0]].describe()[7])
print('TOT_L2:','%.3f' % data[attrs_n[1]].describe()[1],'%.3f' % data[attrs_n[1]].describe()[2],'%.3f' % data[attrs_n[1]].describe()[3],'%.3f' % data[attrs_n[1]].describe()[7])

data[attrs_n].describe()
```

Utilize os atributos 'ENSINO MEDIO' e 'ANOS FINAIS'. Vamos transformá-los por meio da função  logaritmica. Para isso:

```python
print('Resposta: (b) 0.668')
```

```python
#Transformação

#copiando a base 
data_copy = data.copy()

#atribuindo NaN em todos os atributos iguais a zero
data.loc[data['ENSINO MEDIO'] == 0,'ENSINO MEDIO'] = np.nan
data.loc[data['ANOS FINAIS'] == 0,'ANOS FINAIS'] = np.nan
data.head()
```

```python
#atributos
attrs = ['ENSINO MEDIO','ANOS FINAIS']
print('linhas, colunas:',data.shape)
data.boxplot(attrs)
```

```python
#tranformando dados com raiz quadrada
x_MEDIO = np.array(data['ENSINO MEDIO'])
x_FINAIS = np.array(data['ANOS FINAIS'])

#aplicando transformação da raiz quadrada e adicionando na base de dados
data['sqrt(ENSINO MEDIO)'] = np.sqrt(x_MEDIO)
data['sqrt(ANOS FINAIS)'] = np.sqrt(x_FINAIS)

data.head()
```

```python
#removendo linhas com NaN nos atributos selecionados
data = data.dropna(subset=['sqrt(ENSINO MEDIO)'])
data = data.dropna(subset=['sqrt(ANOS FINAIS)'])
```

```python
#regressão linear
from sklearn.linear_model import LinearRegression

x_regression = data['sqrt(ENSINO MEDIO)']
y_regression = data['sqrt(ANOS FINAIS)']

#aqui se faz necessário utilizar o método reshape para formatar o dado em array 2D, pois o linearRegression não 
#aceita array 1D. https://stackoverflow.com/questions/51150153/valueerror-expected-2d-array-got-1d-array-instead
x_regression_np = x_regression.to_numpy().reshape(-1, 1)
y_regression_np = y_regression.to_numpy().reshape(-1, 1)

#aplicando LinearRegression
lm = LinearRegression().fit(x_regression_np,y_regression_np)

#imprimindo coeficiente linear
print('Coeficiente Linear coef_:', '%.3f' % lm.coef_[0][0])
```

```python
#base de dados anterior à questão 4
attrs = ['MUN','TOT_5I']
data_copy.head()
```

```python
#tratando MUN que é categórica nominal
#para a categórica nominal não há relação entre os elementos, por isso é utilizado ordem alfabética para codificar
data_copy['MUN_cod'] = data_copy['MUN'].astype("category").cat.codes
data_copy.head()
```

```python
#tratando TOT_5I que é categórica ordinal

#obtendo as categorias pelos seus valores únicos e a codificação numérica em ordem crescente
categoria_TOT_5I = data_copy['TOT_5I'].unique().sort_values()
numero_TOT_5I = np.arange(data_copy['TOT_5I'].unique().shape[0])

print('Categorias:',categoria_TOT_5I)
print()
print('Codificação:',numero_TOT_5I)
print()
print()

#montando uma tupla de valores com categoria e codificação e jogando em um dicionario
map_TOT_5I = dict(zip(categoria_TOT_5I,numero_TOT_5I))
print('Dicionário:',map_TOT_5I)
```

```python
#atribuindo os valores na base de dados
data_copy['TOT_5I_ord'] = data_copy['TOT_5I'].map(map_TOT_5I)

data_copy.head()
```
