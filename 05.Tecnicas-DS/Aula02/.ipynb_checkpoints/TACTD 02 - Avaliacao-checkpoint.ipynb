Vamos analisar possíveis outliers. Utilize o método da análise da dispersão pelo desvio padrão e inspecione as colunas 'gid', 'Censo', 'PIB', 'Pop_est_2009', 'desemprego', procurando por outliers globais com critério de 2 desvios padrões, i.e. $2\sigma$.

Quantos outliers foram encontrados, respectivamente, para 'gid', 'Censo', 'PIB', 'Pop_est_2009' e 'desemprego'?

(a) 0, 0, 1, 5, 27<br>
(b) 0, 5, 9, 44, 1 <br>
(c) 0, 5, 5, 27, 1 <br>
(d) 0, 5, 5, 44, 2 <br>

```python
#gid
desvp_GID = data['gid'].std()
media_GID = data['gid'].mean()
dataout_std_GID = data[(data['gid'] < media_GID-(desvp_GID*2)) | (data['gid'] > media_GID+(desvp_GID*2))]

#Censo
desvp_CEN = data['Censo'].std()
media_CEN = data['Censo'].mean()
dataout_std_CEN = data[(data['Censo'] < media_CEN-(desvp_CEN*2)) | (data['Censo'] > media_CEN+(desvp_CEN*2))]

#PIB
desvp_PIB = data['PIB'].std()
media_PIB = data['PIB'].mean()
dataout_std_PIB = data[(data['PIB'] < media_PIB-(desvp_PIB*2)) | (data['PIB'] > media_PIB+(desvp_PIB*2))]

#Pop_est_2009
desvp_POP = data['Pop_est_2009'].std()
media_POP = data['Pop_est_2009'].mean()
dataout_std_POP = data[(data['Pop_est_2009'] < media_POP-(desvp_POP*2)) | (data['Pop_est_2009'] > media_POP+(desvp_POP*2))]

#desemprego
desvp_DES = data['desemprego'].std()
media_DES = data['desemprego'].mean()
dataout_std_DES = data[(data['desemprego'] < media_DES-(desvp_DES*2)) | (data['desemprego'] > media_DES+(desvp_DES*2))]

print('Resposta: b)',len(dataout_std_GID),',',len(dataout_std_CEN),',',len(dataout_std_PIB),',',len(dataout_std_POP),',',len(dataout_std_DES))
```

### Exercício 7)

Codifique uma funcão que preencha valores faltantes utilizando a média condicionada a uma outra variável da base. Essa funcao deverá:
1. calcular a média da variável alvo (a ser preenchida) relativa a cada valor distinto da variável selecionada
2. atribuir a média calculada a todas as linhas cuja variavel alvo é faltante e que possua o valor da variável correspondente
3. o valor atribuido deve seguir o mesmo tipo da variável alvo, ou seja, int, float, etc. Quando int, realize o arredondamento utlizando `np.round(,0)`, quando float64 utilize `np.round(value,1)`

Use a funcão para preencher dados faltantes de desemprego condicionado a UF. Considerando arredondamento para 4 casas decimais, qual é a média da coluna desemprego para todas as linhas, antes e depois de realizar o preenchimento?

(a) antes: 6.6406, depois: 6.6423<br>
(b) antes: 6.6406, depois: 6.6406<br>
(c) antes: 6.6423, depois: 6.6406<br>
(d) antes: 6.6423, depois: 6.6423<br>


```python
print('Resposta: a) antes: 6.6406, depois: 6.6423')
```

```python
def preenche_faltantes_com_media(data_origin, var, var_condition):
    
    data = data_origin.copy() #cria uma cópia do database original
    unique_var_condition = data[var_condition].unique() #encontra os valores únicos do atributo condicional
    data_type = data.dtypes[var] #encontra qual o tipo do atributo original
    
    for i in unique_var_condition:
        val = data[data[var_condition] == i][var] #pega os valores dos atributos dada o atributo condicional
        val_mean = np.nanmean(val.astype(float)) #calcula a média ignorando as ocorrências de NaN, obs: aqui é necessário converter para float para calcular a média de valores que não são do tipo float
        val_fill = val.loc[val.isnull() == True].fillna(val_mean).astype(data_type) #preenche os valores com a média convertendo para o formato do atributo
        
        for j in val_fill.index:
            data.loc[j,var] = val_fill[j] #completa o dataframe com os valores preenchidos usando o índice
    
    if((data[var].dtypes == 'int16') | (data[var].dtypes == 'int32') | (data[var].dtypes == 'int64')):
        data[var] = np.round(data[var],0) #se for tipo inteiro arredonda com 0
        
    if ((data[var].dtypes == 'float32') | (data[var].dtypes == 'float64')):
        data[var] = np.round(data[var],1) #se for tipo float arredonda com 1
            
    return data #retorna o novo dataframe com a coluna preenchida
```

```python
data_convert_des = data_dropped
data_convert_des.loc[:,'desemprego'] = pd.to_numeric(data_convert_des.loc[:,'desemprego'], downcast='float', errors='coerce')

ac_mean = data_convert_des['desemprego'].mean()
data_fill_desempregro = preenche_faltantes_com_media(data_convert_des, 'desemprego','UF')

print('Média antes do preenchimento: %.4f' % ac_mean)
print('Média após o preenchimento: %.4f' % data_fill_desempregro['desemprego'].mean())
```

```python
#testando para outros atributos
print('Testando atributos de outros tipos em relação a UF:')
print()
data_fill_classe = preenche_faltantes_com_media(data_convert_des, 'classe','UF')
print('Média Classe: %.4f' % data_fill_classe['classe'].mean()," - Tipo:",data_fill_classe['classe'].dtype)

data_fill_gid = preenche_faltantes_com_media(data_convert_des, 'gid','UF')
print('Média GID: %.4f' % data_fill_gid['gid'].mean()," - Tipo:",data_fill_gid['gid'].dtype)

data_fill_pib = preenche_faltantes_com_media(data_convert_des, 'PIB','UF')
print('Média PIB: %.4f' % data_fill_pib['PIB'].mean()," - Tipo:",data_fill_pib['PIB'].dtype)
```
