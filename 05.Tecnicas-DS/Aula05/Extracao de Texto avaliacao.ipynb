### Exercício 1)
Considere a nota fiscal eletrônica representada no arquivo <font style="font-family: monaco"> nfe-avaliacao.pdf</font>. Converta o arquivo PDF em uma imagem no formato PNG. Utilize o pacote <font color='blue'>PIL</font> para carregar a imagem gerada. Qual a resolução da imagem gerada?

a) 1653 X 2339 <br>
b) 1024 X 640 <br>
c) 640 X 2339<br>
d) 1001 X 2020


**Dica**: Utilize o atribute `size` do objeto PIL para obter as dimenões da imagem e empregue os parâmetros default do método <font color='blue'>convert_from_path</font> para gerar a imagem.


```python
from pdf2image import convert_from_path 
from PIL import Image

imagem = convert_from_path('.\img\\nfe-avaliacao.pdf')
for i, pagina in enumerate(imagem):
    filename = './img/paginasAvaliacao/pagina_'+str(i)+'.png'
    pagina.save(filename,'PNG')

Image.open('.\img\paginasAvaliacao\pagina_0.png').size

print('Resposta:',Image.open('.\img\paginasAvaliacao\pagina_0.png').size)
```

### Exercício 2)
Aplique OCR para extrair o texto contido na imagem gerada no exercício 1). Quantas linhas o texto resultante possui?

a) 230<br>
b) 231<br>
c) 232<br>
d) 233

**Dica**: Para contar as linhas, procure pelo número de ocorrências do símbolo '\n' no texto extraído.

```python
import pytesseract as ocr
from PIL import Image
import glob
import re

files = glob.glob('./img/paginasAvaliacao/*') #glob é um filename pattern matching, serve para encontrar arquivos

texto=[]
count = 0

for filename in files: #pega todos os arquivos encontrados e converte a imagem para string com OCR, abrindo pelo PIL
    t = ocr.image_to_string(Image.open('.\img\paginasAvaliacao\pagina_0.png'),lang='por')
    texto.append(t)

for i,t in enumerate(texto): #i neste caso recebe o número de páginas do texto, t contém o texto convertido
    texto_integro = texto[i] #imprimindo as paginas encontradas numa variável texto

for i in texto_integro: #leio o texto, onde i tem cada caractere do texto 
    if i == '\n': #se i for uma linha nova (\n) eu conto.
        count = count + 1
        
print("Resposta:",count)
print()
print(texto_integro)

```

### Exercício 3)
Escreva uma expressão regular para encontrar todos os valores financeiros descritos na nota. Ou seja, todos as ocorrências de uma sequência de dígitos que, precedem uma vírgula, a qual é seguida de exatamente outros dois dígitos (por exemplo: 7545,43). Quantas ocorrências de valores financeiros existem na NEF?


a) 32<br>
b) 33<br>
c) 36<br>
d) 38

```python
import re

valores = re.findall('[\d]*,[\d]+',texto_integro) #regex para encontrar os valores financeiros

valInvalido = re.findall('[\d]*,[\d]{3}',str(valores)) #regex para encontrar valores inválidos (com mais de 3 casas decimais)

print('Valores inválidos:')
print(valInvalido)
print()
print('Valores obtidos:')
print(valores)
print()

#percorre a lista de valores obtidos X vezes (X é o tamanho do array de valores inválidos)
#remove o primeiro valor inválido encontrado e vai repetindo X vezes, até não termos mais valores inválidos
for i in range(len(valInvalido)):
    valores.remove(re.findall('[\d]*,[\d]{3}',str(valores))[0]) 
    
print('Removendo valores inválidos:')
print(valores)
print()
print("Resposta:",len(valores))
```

### Exercício 4)
Dos valores financeiros obtidos no exercício anterior, os de valor maior que R$ 100,00 somam:

a) 17456.65<br>
b) 19122.95<br>
c) 36456.25<br>
d) 20345.86

**Dica**: Converta as strings para float.

```python
val=[]
soma=0.0

for i in valores:
    val.append(float(re.sub(",",".",i))) #regex para substituir "," por "." para poder converter para float

print('Valores maiores que 100:')
for j in val:
    if (j >= 100): #procurando valores maiores que 100
        print(j)
        soma = soma+j # somando valores maiores que 100
print()
print('Resposta:',soma)
```
