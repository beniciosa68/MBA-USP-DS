### Questão 1 (2.5 pontos)
Considere o arquivo <font style="font-family: monaco"> modcovid.pdf </font> (disponível para download no moodle). Escreva um código para extrair o texto (ASCII) do arquivo PDF e escreva o texto extraído em um arquivo chamado ``modcovid.txt``.


```python
import glob
from PIL import Image
import pytesseract as ocr
from pdf2image import convert_from_path 

paginas = convert_from_path('.\data\modcovid.pdf') #obtendo páginas do PDF

for i,pag in enumerate(paginas):
    file = './data/modcovid/pagina_'+str(i)+'.PNG' #nomeando páginas
    pag.save(file,'PNG') #salvando páginas em imagem PNG

files = glob.glob('./data/modcovid/*') #verificando conteúdo percorrendo arquivos e diretórios

texto=[]

for f in files:
    t = ocr.image_to_string(Image.open(f),lang='por') #ocr para converter imagem para texto
    texto.append(t)

    
arquivo = open('.\data\modcovid.txt','wt') #abrindo arquivo para escrita

for i,t in enumerate(texto):
    n = arquivo.write(texto[i]) #escrevendo em arquivo modcovid.txt
    
arquivo.close()
```

### Questão 2 (2.5 pontos)
Leia o arquivo ``modcovid.txt`` e realize as seguinte operações:
1. Extraia todas palavras contidas no arquivo e armazene em uma lista de palavras (utilize o método <font color='blue'>word_tokenize</font> do pacote <font color='blue'>nltk</font>.
2. Remova da lista de palavras todos os "palavras" que não sejam formadas exclusivamente de caracteres do alfabeto.
3. Quantas palavras com apenas 1 caractere sobraram na lista?

```python
import nltk

arquivo = open('.\data\modcovid.txt','r') #lendo o arquivo

#item 1
docs = '' #string com palavras do arquivo
for w in arquivo:
        docs = docs+' '+w

palavras = nltk.word_tokenize(docs) #transformando a string em lista de palavras

print('Lista de palavras:')
print()
print(palavras)
```

```python
#item 2
palavras = [p.lower() for p in palavras if p.isalpha()] #retorna strings formadas apenas por letras do alfabeto

print(palavras)
```

```python
#item 3
c=0
for p in palavras:
    if len(p) == 1:
        c = c + 1 #contando as palavras com tamanho 1

print('Temos',c,'palavras com apenas 1 caractere na lista.')
```

#### b) (1 ponto)

Execute uma função que identifique outliers com base no intervalo interquartil.
Mostre as linhas referentes a outliers detectados por esse método no atributo `paintings` para valores para além de mais ou menos $2 \times IQR$.

```python
def detect_outlier(data,feature): #criando uma função para detectar outliers usando o método do IQR
    
    Q1 = data[feature].quantile(0.25) #primeiro quartil
    Q3 = data[feature].quantile(0.75) #terceiro quartil
    IQR = Q3 - Q1 #calculando IQR
    dataout_iqr = data[(data[feature] < Q1 - (IQR * 2)) 
                    |  (data[feature] > Q3 + (IQR * 2))] #calculando outlier considerando intervalo de mais ou menos 2 x IQR

    return dataout_iqr

detect_outlier(df,'paintings')
```

### Questão 4 (2.5 pontos)

Dada uma imagem `painting.jpg` de uma pintura da qual não sabemos o artista, gostaríamos de fazer uma busca numa base de dados e recuperar obras similares. Para isso utilizaremos uma composição de descritores:
1. Histograma de cores (R, G, B) considerando 4 bins por canal de cor (total 12 características)
2. Descritor LBP utilizando raio 2.5 e 16 pontos (total 18 características).

Concatene esses dois descritores e use-o como descritor de cor e textura da imagem. Faça uma busca no diretório `paintings`, retornando as 5 imagens mais similares de acordo com esse descritor e a distância Euclidiana. Exiba a imagem de consulta e também as 5 imagens retornadas, com seus nomes e valor da distância obtido.

```python
# inclua os pacotes necessários e as funções necessárias
import glob
import imageio
import matplotlib
import numpy as np
import pandas as pd
from skimage import data
from skimage import feature
import matplotlib.pyplot as plt

#implementando a função de histograma de cores
def histograma_cores(img, bins): 
    
    if(len(img.shape) == 2): #para um canal de cor (preto e branco)
        hist,_ = np.histogram(img, bins=bins)
        
    if(len(img.shape) == 3): #para mais que um canal de cor (R,G,B)
        hist_R,_ = np.histogram(img[:,:,0], bins=bins) #canal vermelho
        hist_G,_ = np.histogram(img[:,:,1], bins=bins) #canal verde
        hist_B,_ = np.histogram(img[:,:,2], bins=bins) #canal azul

        hist = np.concatenate([hist_R, hist_G, hist_B])
        
    hist = hist.astype("float")
    hist /= (hist.sum() + 0.0001) #normalização        
    
    return hist

#implementando a função do descritor lbp
def lbp_features(img, points=16, radius=2.5):
    
    img = np.array(img, dtype=np.float64, copy=False) #criando um array com os dados da imagem
    
    if(len(img.shape) > 2): #lbp trabalha com apenas 1 canal de cor, se for RGB faz a conversão com o método Luminance
        img = img[:,:,0]*0.3 + img[:,:,1]*0.59 + img[:,:,2]*0.11
        
    lbp = feature.local_binary_pattern(img.astype(np.uint8), points, radius, method="uniform") #caracteristicas points e radius
    
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, points+3), range=(0, points+2)) #extraindo histograma da matriz lbp
    
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-6) #normalização
    
    return hist

#implementando a função do descritor de cor e textura de imagem
def histograma_cores_textura(img, bins=4, points=16, radius=2.5):
    
    hist_cores = histograma_cores(img,bins) #aplicando histograma de cores
    hist_textura = lbp_features(img, points, radius) #aplicando histograma lbp para textura
    hist = np.concatenate([hist_cores, hist_textura]) #concatenando histogramas
    
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-6) #normalização
    
    return hist

#implementando a função de cálculo da distância euclidiana
def distancia_euclidiana(A, B): 
    return  np.sqrt( np.sum((A-B)**2) ) #calcula a distancia euclidiana entre dois vetores de mesmo tamanho
```

```python
# inclua o código para carregar as imagens, gerar os vetores de características e obter as distâncias

paintings = glob.glob('./data/paintings/*') #verificando conteúdo percorrendo arquivos e diretórios

img_ref = imageio.imread('./data/paintings\\painting.jpg') #pegando a imagem de referência

img=[]
for p in paintings:
    if(p != './data/paintings\\painting.jpg'): #removendo a imagem de referência, para não interferir no cálculo da distância
        img.append(imageio.imread(p)) #percorrendo as imagens de comparação

fig, axes = plt.subplots(6, 10, figsize=(18, 15))
ax = axes.ravel()

for i in range(len(img)):
    ax[i].imshow(img[i], cmap=plt.cm.gray)
    ax[i].axis('off')
fig.tight_layout() #imprimindo as imagens de comparação, exceto a imagem de referência
```

```python
# inclua o código para obter as 5 imagens mais próximas com base nas distâncias computadas e exibi-las
dE = []

for i in range(len(img)):
        dE.append([distancia_euclidiana(histograma_cores_textura(img_ref),histograma_cores_textura(img[i]))])
        
data = pd.DataFrame(dE,columns=['rank_dist_img_ref']) #inserindo os dados dos cálculos em um df

print(data.sort_values(by='rank_dist_img_ref',ascending=True)[:5]) #imprimindo top 5 com respectivos índices
print()
```
