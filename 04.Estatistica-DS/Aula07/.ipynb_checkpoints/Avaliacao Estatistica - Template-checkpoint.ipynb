````python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols

data = pd.read_csv('.\data\Prestigio.csv', sep=',', na_values=',',encoding='unicode_escape')
data = data.dropna()
data = pd.concat([data, pd.get_dummies(data['type'].astype('category'), prefix = 'd')], axis = 1) 
print(data.head())
data.describe()
````

````python
#modelo completo:
income_c = data.income-np.mean(data.income)
education_c = data.education-np.mean(data.education)
women_c = data.women-np.mean(data.women)

mod = ols('data.prestige ~ income_c + education_c + data.d_prof + data.d_wc + women_c',data=data)
res = mod.fit()
print(res.summary())
````

````python
#Realizando a análise de backward através do p-valor:
print('Para um nível de significância alpha de 5%, temos que o p-valor de women não encontra-se no intervalo')
print('Removendo:')
print('women_c p-valor:0.832 que não está no intervalo de confiança: -0.054, 0.067')

#modelo sem women_c:
mod = ols('data.prestige ~ income_c + education_c + data.d_prof + data.d_wc',data=data)
res = mod.fit()
print(res.summary())
````

````python
#Realizando a análise de backward através do p-valor:
print('Para um nível de significância alpha de 5%, temos que o p-valor de wc não encontra-se no intervalo IC para 5%')
print('Removendo:')
print('data.d_wc p-valor: 0.279 não está no IC: -7.729, 2.255')

#modelo sem women_c:
mod = ols('data.prestige ~ income_c + education_c + data.d_prof',data=data)
res = mod.fit()
print(res.summary())
````
