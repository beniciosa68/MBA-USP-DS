```python
#lendo arquivos com letras de músicas (exemplo: top 10 musicas nacionais do site letras.com.br)
import glob
import nltk
files = glob.glob("musicas/top10-nacional/*")
print(files)

docs = ''
for fname in files:
    with open(fname,'r') as f:
        docs = docs +' '+ f.read()
```

```python
#dividindo todas as letras de músicas e separando em palavras
words = nltk.word_tokenize(docs)
print(words)
```

```python
#removendo símbolos e caracteres
import string
words = [w.lower() for w in words if w.isalpha() and w not in string.ascii_letters]
print(words)
```

```python
#removendo stop words (palavras que não são úteis para trazer algum significado, não contribuem para a semântica)
from nltk.corpus import stopwords

stop_words = stopwords.words('portuguese')

words = [w for w in words if w not in stop_words]

print(words)
```

```python
#bigramas, trigramas...n-gramas (grupo de duas ou mais palavras que tem grande significado semântico)
from nltk import word_tokenize
from nltk.collocations import *

#medidas de importância para n-gramas
bigram_measures = nltk.collocations.BigramAssocMeasures()

#encontrando bigramas
finder = BigramCollocationFinder.from_words(words)

for i in finder.nbest(bigram_measures.raw_freq,20):
    words.append(i[0].lower() + ' ' + i[1].lower())

print(words[-50:])
```

```python
#normalização léxica
#stemming (converter sua palavra para sua forma base -> andando, andarei, andares -> andar)
from nltk.stem import PorterStemmer

words = [PorterStemmer().stem(w) for w in words]

print(words)
```

```python
#lematização (método que leva em consideração o texto como um todo pra fazer a redução léxica baseado na sequencia de palavras)
from nltk.stem.wordnet import WordNetLemmatizer

words = [WordNetLemmatizer().lemmatize(w,'v') for w in words]

print(words)
```
