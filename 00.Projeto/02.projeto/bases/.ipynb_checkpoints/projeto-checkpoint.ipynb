```python
import pandas as pd
from pylab import rcParams
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from statsmodels.tsa.seasonal import seasonal_decompose

import seaborn as sns
from statsmodels.formula.api import ols
from scipy.stats import probplot
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
```

```python
df = pd.read_csv('sgemm_product.csv',sep=',')
print(df.shape)
df.head()
```

#### Descrição dos dados
    - Input:
        MWG, NWG: per-matrix 2D tiling at workgroup level: {16, 32, 64, 128} (integer)
        KWG: inner dimension of 2D tiling at workgroup level: {16, 32} (integer)
        MDIMC, NDIMC: local workgroup size: {8, 16, 32} (integer)
        MDIMA, NDIMB: local memory shape: {8, 16, 32} (integer)
        KWI: kernel loop unrolling factor: {2, 8} (integer)
        VWM, VWN: per-matrix vector widths for loading and storing: {1, 2, 4, 8} (integer)
        STRM, STRN: enable stride for accessing off-chip memory within a single thread: {0, 1} (categorical)
        SA, SB: per-matrix manual caching of the 2D workgroup tile: {0, 1} (categorical)
    
    - Output:
        Run1, Run2, Run3, Run4: performance times in milliseconds for 4 independent runs using the same parameters.

```python
#detectando e removendo outliers
def detect_remove_outlier(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3-Q1
    mask = ((data > (Q1 - 1.5 * IQR)) & (data < (Q3 + 1.5 * IQR)))
    data_clean = data[mask]
    return data_clean

df_clean = detect_remove_outlier(df)
df_clean.head()
```

```python
#regressão linear simples

mod1 = ols('rtt ~ worker * dimatrix + worker * nbytes + dimatrix * nbytes',data=df)
res1 = mod1.fit()
print(res1.summary())
```

```python
mod2 = ols('rtt ~ dimatrix + worker',data=df)
res2 = mod2.fit()
print(res2.summary())
```

###### Indicação de multicolinearidade dos dados

Voltando com as suspeitas de multicolinearidade identificadas no início da análise, iremos utilizar o cálculo do fator de inflação da variância (VIF) para as variáveis explicativas do modelo. O critério a ser utilizado para análise do VIF (inflação de variância) será:

VIF for igual à 1 não há multicolinearidade entre os fatores;

VIF acima de 1, as preditoras podem estar correlacionadas.

De 1 até 5: indica alguma correlação, porém, não o suficiente para impactar no modelo;
De 5 até 10: alta correlação podendo gerar impacto no modelo;
Acima de 10: coeficientes de regressão estão mal estimados devidos à multicolinearidade;

```python
var = res2.model.exog
vif = [variance_inflation_factor(var, i) for i in range(var.shape[1])]

print('Intercept:',vif[0])
print('dimatrix:',vif[1])
print('worker:',vif[2])
```

```python
#padronização (variável menos média dividido pelo desvio padrão)
dimatrix_p = (df.dimatrix-np.mean(df.dimatrix))/(np.std(df.dimatrix))

mod3 = ols('rtt ~ dimatrix_p + worker',data=df)
res3 = mod3.fit()
print(res3.summary())
```

```python
var = res3.model.exog
vif = [variance_inflation_factor(var, i) for i in range(var.shape[1])]

print('Intercept:',vif[0])
print('dimatrix_p:',vif[1])
print('worker:',vif[2])
```
